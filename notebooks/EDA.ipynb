Exploring Data 

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

```python
df['Dt_Customer']
```

```python
#Dropping ID column 

columns = [column for column in df.columns if column!="ID"]
```

```python
#Categorizing Numerical and Categorical Features

numeric_features = [feature for feature in columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in columns if df[feature].dtype == 'O']

print('There are {} numerical features: {}'.format(len(numeric_features), numeric_features))
print('\n\nThere are {} categorical features: {}'.format(len(categorical_features), categorical_features))
```

```python
#Count Unique values of Education column

df[categorical_features[0]].value_counts().count()
```

```python
#Percentage of values of each categorical columns 
for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('\n------------------------------------------\n')
```

Univariate Analysis of Numerical Features 

```python
plt.figure(figsize = (15,10))
plt.suptitle("Univariate Analysis of Numerical Features", fontsize=20, fontweight='bold', alpha=0.8, y=1)

for i in range(0, len(numeric_features)):
    plt.subplot(7, 4, i+1)
    sns.kdeplot(x=df[numeric_features[i]], color='blue')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()


plt.show()
```

```python
df.Z_CostContact.plot()
plt.show()
```

Both Z_Revenue and Z_CostContact columns have constant value. Since there is no variance in the column therefore we will drop these 2 columns. 

```python
df = df.drop(columns=['Z_CostContact', 'Z_Revenue'])
df.columns
```

Categorical Features

```python
plt.figure(figsize=(15,8))
plt.suptitle("Univariate Analysis of Categorical features", fontsize=20, fontweight='bold', alpha=0.8, y=1)

for i in range(0, len(categorical_features)):
    plt.subplot(2, 2, i+1)
    ax=sns.countplot(x=df[categorical_features[i]])
    if df[categorical_features[i]].value_counts().count() < 10:
        for p in ax.patches:
            ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
    plt.xlabel(categorical_features[i])
    plt.tight_layout()


plt.show()
```
