Exploring Data 

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

```python
df['Dt_Customer']
```

```python
#Dropping ID column 

columns = [column for column in df.columns if column!="ID"]
```

```python
#Categorizing Numerical and Categorical Features

numeric_features = [feature for feature in columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in columns if df[feature].dtype == 'O']

print('There are {} numerical features: {}'.format(len(numeric_features), numeric_features))
print('\n\nThere are {} categorical features: {}'.format(len(categorical_features), categorical_features))
```

```python
#Count Unique values of Education column

df[categorical_features[0]].value_counts().count()
```

```python
#Percentage of values of each categorical columns 
for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('\n------------------------------------------\n')
```
